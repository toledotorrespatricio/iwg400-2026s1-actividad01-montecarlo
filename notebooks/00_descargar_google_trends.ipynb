# 00 Descarga de Google Trends

Este notebook descarga datos **reales** de Google Trends para evitar trabajar con un CSV inventado.

Usaremos `pytrends`, que es una librería **no oficial**. Si falla, el plan B es exportar manualmente desde la web de Trends.

## Qué haremos

1. Importar la librería que conversa con Google Trends.
2. Definir términos, país y ventana temporal.
3. Pedir la tabla de interés en el tiempo.
4. Guardar el resultado en `../data/google_trends.csv`.

## Ojo

Google no ofrece una API pública oficial sencilla para esto. Por eso este paso puede dejar de funcionar si Google cambia su servicio.

```python
from pathlib import Path

import pandas as pd
from pytrends.request import TrendReq

# Path nos ayuda a construir rutas de archivos sin preocuparnos por barras de Windows o Linux.
# pandas nos sirve para trabajar con tablas.
# TrendReq es la clase de pytrends que hace la consulta a Google Trends.
```

```python
# Elegimos los términos que queremos comparar.
keywords = ["ChatGPT", "TikTok", "Instagram"]

# geo = "CL" significa Chile.
geo = "CL"

# timeframe = "today 3-m" significa últimos 3 meses.
timeframe = "today 3-m"

# Creamos la conexión con Google Trends.
pytrends = TrendReq(hl="es-CL", tz=180)

# build_payload prepara la consulta. Todavía no descarga la tabla.
pytrends.build_payload(keywords, cat=0, timeframe=timeframe, geo=geo, gprop="")
```

```python
# interest_over_time() pide la serie temporal.
df = pytrends.interest_over_time()

# A veces Google agrega la columna isPartial. Para esta actividad no la necesitamos.
if "isPartial" in df.columns:
    df = df.drop(columns=["isPartial"])

# Convertimos la fecha desde indice a columna normal.
df = df.reset_index()

df.head()
```

```python
# Guardamos el archivo para que los otros notebooks puedan leerlo.
output = Path("../data/google_trends.csv")
output.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(output, index=False)

print(f"Archivo guardado en: {output.resolve()}")
print(f"Filas descargadas: {len(df)}")
print("Columnas:", list(df.columns))
```

## Si esto falla

- prueba otra ventana temporal, por ejemplo `today 12-m`;
- espera unos minutos y vuelve a intentar;
