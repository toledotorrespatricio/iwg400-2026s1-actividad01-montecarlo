En este notebook vamos a abrir una tabla real de Google Trends y responder preguntas simples con `pandas` y `matplotlib`.

## Definiciones simples

- **CSV**: archivo de texto que guarda tablas.
- **DataFrame**: tabla que `pandas` usa dentro de Python.
- **Serie temporal**: datos ordenados por fecha.
- **Promedio**: valor medio de una columna.
- **Máximo**: el valor más alto observado.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

# matplotlib sirve para dibujar gráficos.
# pandas sirve para leer la tabla y calcular resúmenes.
# Path nos ayuda a ubicar el archivo.
```

```python
# Definimos dónde debería estar el archivo descargado desde Google Trends.
path_csv = Path("../data/google_trends.csv")

# Verificamos que exista antes de seguir.
if not path_csv.exists():
    raise FileNotFoundError(
        "No existe ../data/google_trends.csv. Ejecuta primero 00_descargar_google_trends.ipynb o scripts/fetch_google_trends.py"
    )

# Leemos el CSV y convertimos la columna date a tipo fecha.
df = pd.read_csv(path_csv)
df["date"] = pd.to_datetime(df["date"])

# Mostramos las primeras filas para entender la estructura.
df.head()
```

```python
# Miramos nombres de columnas, tipos y tamaño de la tabla.
print("Columnas:", list(df.columns))
print()
print(df.info())
```

```python
# Identificamos qué columnas contienen datos de interés y no fechas.
terminos = [col for col in df.columns if col != "date"]

# Calculamos promedio y máximo por término.
resumen = pd.DataFrame({
    "promedio": df[terminos].mean(),
    "maximo": df[terminos].max(),
})

resumen.sort_values("promedio", ascending=False)
```

```python
# Buscamos la fecha exacta en que cada término alcanzó su valor máximo.
fechas_pico = {}
for termino in terminos:
    fila_max = df.loc[df[termino].idxmax()]
    fechas_pico[termino] = {
        "valor_maximo": int(fila_max[termino]),
        "fecha_pico": fila_max["date"].date().isoformat(),
    }

pd.DataFrame(fechas_pico).T
```

```python
# Dibujamos las series para comparar cómo cambia el interés en el tiempo.
plt.figure(figsize=(10, 5))

for termino in terminos:
    # Cada término se dibuja como una línea distinta.
    plt.plot(df["date"], df[termino], marker="o", linewidth=2, label=termino)

plt.title("Interés relativo en Google Trends")
plt.xlabel("Fecha")
plt.ylabel("Interés")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
